```python
########## Lambda 

from functools import reduce
p = 0
li = [0,1]
while p < 15:
    
    sum = str(reduce((lambda x,y : x + y),li[-2:]))
    li.append(int(sum))
    p += 1
else:
    print(li)
```

```python
######### Lambda 1

from functools import reduce    
fb = lambda n: reduce(lambda x,_: x+[x[-1]+x[-2]], range(n-2),[0,1])
fb(15)
```

```python
########## vinay code for web scraping and reading the json file


import urllib.request, json
import urllib.parse

city = input("Enter city name")
url = "http://api.openweathermap.org/data/2.5/weather?appid=b35975e18dc93725acb092f7272cc6b8&q="+city
#query_args = {'q': city, 'APPID': 'b35975e18dc93725acb092f7272cc6b8'}

#data = urllib.parse.urlencode(query_args).encode('utf-8')

req = urllib.request.Request(url)

res = urllib.request.urlopen(req)
data = res.read()

dData = data.decode('utf-8')
print(dData)
jsonData = json.loads(dData)

main = jsonData["main"]                                             ######extracting the JSON data 
wind = jsonData["wind"]
weather = jsonData["weather"]
clouds = jsonData["clouds"]
print("Today's temperature is :"+str(main['temp']))
print("Today's cloud is :"+str(clouds['all']))
print("Wind speed :"+str(wind['speed']))
for item in weather:
    print("Description : "+item['description'])
    print("Weather :"+item['main'])
```
